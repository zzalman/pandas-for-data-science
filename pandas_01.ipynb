Bining

now we are moving towrds bining. bining is most important in data analysis
in this we make a new columns of category based on numerical columns makes groups basically

```python
df.age.min()
```

```python
df.age.max()
```

```python
sns.histplot(df.age)
```

```python
# histogram are made to see the distribution of data
# only made up for numerical columns
# if the shape is bell curve then it is normally distributed
# if the shape is skewed left or right then it is skewed distribution
# the other names of normally distributed data are
# Gaussian distribution and
# Normal distribution
# Bell curve distribution
# and mostly the normally distributed data are when data is in large form and correctly measured

```

```python
# now we see how bining works in pandas
# bining age columns into 7 categories
bins = [0,1,5,12,18,30,50,80]
# now names these categories as a labels
labels = ['infrances','toodlers','kids','teens','youngs','middle age','old']
# now make sclicing 
pd.cut(df.age, bins=bins, labels=labels)
```

```python
# now you can assign this to a new column if you want
df['age_group'] = pd.cut(df.age, bins=bins, labels=labels)
df.head()
```

```python
# adding in column age_group is basically feature engineering and it can help in analysis and modeling
# feature engineering is the most important part of data science and machine learning
# feature engineering is the process of using domain knowledge to extract features from raw data via
# various techniques.
# feature engineering can significantly improve the performance of machine learning models.
# feature engineering techniques include:
# - Binning
# - Encoding
# - Scaling
# feature engineering means basically creating new features from existing features to improve model performance.
# features is another name of columns in data science and machine learning terminology.
```

```python
df['age_group'].value_counts()
```

```python
# you just need to think what you want to do with your data and then choose the best method accordingly.
# and this is not just end of feature engineering there are many more techniques to do feature engineering
```

```python
# renaming a column is also part of feature engineering
df.rename(columns={'age_group':'age_category'}, inplace=True)
```
