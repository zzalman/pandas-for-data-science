Size Variable correction

```python
df['Size'].value_counts()
```

varies with devices means null values so we will replace them with nan 

```python
df['Size'] = df['Size'].replace('Varies with device',np.nan)
```

```python
df['Size'].value_counts()
```

```python
df['Size_MB'] = np.nan

kb_mask = df['Size'].str.endswith('k', na=False)
mb_mask = df['Size'].str.endswith('M', na=False)

df.loc[kb_mask, 'Size_MB'] = (
    df.loc[kb_mask, 'Size']
    .str.replace('k', '', regex=False)
    .astype(float) / 1024
)

df.loc[mb_mask, 'Size_MB'] = (
    df.loc[mb_mask, 'Size']
    .str.replace('M', '', regex=False)
    .astype(float)
)
```

```python
df.drop(columns=['Size'], inplace=True)
```

```python
df.rename(columns={'Size_MB':'Size'}, inplace=True)
```

ok so we convert 2 variable so far one remianing for that we can make bins for make categorical type install


```python
df['Installs'].value_counts()
```

```python
df['Installs'] =( df['Installs']
    .str.replace('+', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(int))
```

```python
df['Installs'].unique()
```

```python
bins = [-1, 0, 100, 10_000, 1_000_000, 100_000_000, 1_500_000_000]
labels = [
    'No installs',
    'Very Low',
    'Low',
    'Medium',
    'High',
    'Very High'
]

df['Install_Category'] = pd.cut(
    df['Installs'],
    bins=bins,
    labels=labels
)
```
