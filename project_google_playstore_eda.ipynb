```python
import numpy as np

# Replace text with NaN
df['Size'] = df['Size'].replace('Varies with device', np.nan)

# KB → Bytes
kb_mask = df['Size'].str.endswith('k', na=False)
df.loc[kb_mask, 'Size'] = (
    df.loc[kb_mask, 'Size']
    .str.replace('k', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float) * 1024
)

# MB → Bytes
mb_mask = df['Size'].str.endswith('M', na=False)
df.loc[mb_mask, 'Size'] = (
    df.loc[mb_mask, 'Size']
    .str.replace('M', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float) * 1024*1024
)

# Ensure numeric dtype
df['Size'] = df['Size'].astype(float)
```
