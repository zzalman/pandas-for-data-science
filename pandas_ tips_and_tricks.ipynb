# 26 spliting strings into multipul column


```python
df = pd.DataFrame({'name':['muhammad salman','ali ahmed'],'location':['lahore, pakistan','multan, pakistan']})
df
```

```python
# split the column
df.name.str.split(' ',expand=True)
```

```python
# adding those split columns
df[['first_name','last_name']] = df.name.str.split(' ', expand=True)
df
```

```python
# now expand location
df[['city','country']] = df.location.str.split(",",expand=True)
df
```

```python
# now drop the previous two columns
df.drop(columns=['name','location'])
```

```python
df
```

```python
#now re-arrange columns name 
df =df[['first_name','last_name','name','city','country','location']]
df
```
